```python
#Code to accept the lasio files that only have the required curves
import lasio
import os
import glob

# Required curve list
required_curves = {
    'DEPT', 'AVTX', 'BVTX', 'CILD', 'CNDL', 'CNLS', 'CNPOR', 'CNSS', 'DCAL', 'DPOR',
    'GR', 'LSPD', 'LTEN', 'RHOB', 'RHOC', 'RILD', 'RILM', 'RLL3', 'SP', 'MCAL', 'MI',
    'MN', 'DT', 'ITT', 'SPOR'
}

# Folder path where .las files are stored
las_folder = r"C:\Users\Admin\Downloads\las_files"
las_files = glob.glob(os.path.join(las_folder, "*.las"))

# Output folder for valid individual LAS files
valid_folder = os.path.join(las_folder, "valid_las2")
os.makedirs(valid_folder, exist_ok=True)

# Collect all valid las objects
valid_las_list = []

for las_file in las_files:
    try:
        las = lasio.read(las_file)
        available_curves = {curve.mnemonic.strip().upper() for curve in las.curves}
        
        if required_curves.issubset(available_curves):
            print(f"✅ Accepted: {os.path.basename(las_file)}")

            # Optional: Save valid file for reference
            filename = os.path.basename(las_file)
            las.write(os.path.join(valid_folder, filename))
            valid_las_list.append(las)

        else:
            print(f"❌ Skipped (missing curves): {os.path.basename(las_file)}")

    except Exception as e:
        print(f"⚠️ Error reading {las_file}: {e}")

```

```python
#code to accept lasio files accepting atleast features(required_curves) + other features 
import lasio
import os
import glob

# Minimum required curves
required_curves = {
    'DEPT', 'RHOB', 'RHOC', 'DPOR', 'GR', 'RILM', 'RLL3', 'RILD', 'MN', 'CNLS'
}

# Folder path where .las files are stored
las_folder = r"C:\Users\Admin\Downloads\las_files"
las_files = glob.glob(os.path.join(las_folder, "*.las"))

# Output folder for valid individual LAS files
valid_folder = os.path.join(las_folder, "valid_las3")
os.makedirs(valid_folder, exist_ok=True)

# Collect all valid las objects
valid_las_list = []

# Process each file with progress tracking
for idx, las_file in enumerate(las_files, start=1):
    try:
        las = lasio.read(las_file)
        available_curves = {curve.mnemonic.strip().upper() for curve in las.curves}
        
        if required_curves.issubset(available_curves):  # Accept files with at least the required curves
            print(f"{idx}/{len(las_files)} ✅ Accepted: {os.path.basename(las_file)}")

            # Save valid file
            filename = os.path.basename(las_file)
            las.write(os.path.join(valid_folder, filename))
            valid_las_list.append(las)

        else:
            missing = required_curves - available_curves
            print(f"{idx}/{len(las_files)} ❌ Skipped (missing required curves: {missing}): {os.path.basename(las_file)}")

    except Exception as e:
        print(f"{idx}/{len(las_files)} ⚠️ Error reading {os.path.basename(las_file)}: {e}")
```

```python
#code to integrate all the data of the lasio files in one single csv file

import os, glob, lasio, pandas as pd

LAS_FOLDER = r"C:\Users\Admin\Downloads\las_files\valid_las2"  # <- EDIT THIS
SAVE_CSV   = "all_wells_combined_all_features.csv"

las_paths = glob.glob(os.path.join(LAS_FOLDER, "*.las"))
print(f" Found {len(las_paths)} LAS files")

frames = []
errored_files = []
well_id_to_index = {}
well_index_counter = 0

for path in las_paths:
    try:
        las = lasio.read(path)

        df = las.df().reset_index()
        df.rename(columns={'index': 'DEPT'}, inplace=True)
        df.columns = [c.upper() for c in df.columns]

        well_id = (
            getattr(las.well, "WELL").value.strip()
            if hasattr(las.well, "WELL") and getattr(las.well, "WELL").value
            else os.path.splitext(os.path.basename(path))[0]
        )

        if well_id not in well_id_to_index:
            well_id_to_index[well_id] = well_index_counter
            well_index_counter += 1

        df['WELL_ID'] = well_id
        df['WELL_INDEX'] = well_id_to_index[well_id]

        frames.append(df)

    except Exception as e:
        print(f" Error reading {os.path.basename(path)}: {e}")
        errored_files.append((os.path.basename(path), str(e)))

if not frames:
    print(" No LAS files could be loaded.")
else:
    combined = pd.concat(frames, ignore_index=True, sort=True)
    print(f" Combined shape: {combined.shape} (rows × columns)")

    combined.to_csv(SAVE_CSV, index=False)
    print(f" CSV saved → {SAVE_CSV}")

if errored_files:
    print("\n  Files that could not be read:")
    for fname, err in errored_files:
        print(f"  {fname}: {err}")
```
